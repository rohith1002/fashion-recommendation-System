```python
store=[[]]
for k in comb:
    l1=[]
    j=0
    for i in range(len(k)):
        if(k[i]=='g'):
            l1.append(k[j:i+1])
            j=i+1
    store.append(l1)
```

```python
paths= [r"D:\fcs_dataset\topwear", r"D:\fcs_dataset\bottomwear", r"D:\fcs_dataset\footwear"]
image_paths=[]
for i in range(len(store)):
    
    l=[]
    for j in range(len(store[i])):
        s=''
        s=s+paths[j]+"\\"+store[i][j]
        l.append(s)
    image_paths.append(l)
```

```python
image_paths
```

```python
import cv2


pixels_x1=[]
pixels_x2=[]
pixels_x3=[]
for i in image_paths:
    for j in range(len(i)):
        img=cv2.imread(i[j])
        img=cv2.resize(img,(128,128))
        if(j==0):
            pixels_x1.append(img)
        elif(j==1):
            pixels_x2.append(img)
        else:
            pixels_x3.append(img)
            
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
label_encoder = LabelEncoder()

# Fit and transform the categorical values
encoded_categories = label_encoder.fit_transform(df['Occasion'])

# Convert encoded categories to one-hot encoded format
one_hot_encoded = to_categorical(encoded_categories)

print(one_hot_encoded)
```

```python
pixels_x1=np.array(pixels_x1)
pixels_x2=np.array(pixels_x2)
pixels_x3=np.array(pixels_x3)
```

```python
pixels_x1.shape
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate, BatchNormalization, Dropout

# Define input shapes
image_input_shape = (128, 128, 3)  # Adjust based on your image dimensions
num_occasions = 3  # Number of occasion categories

# Create input layers for images
image_input1 = Input(shape=image_input_shape, name='image_input1')
image_input2 = Input(shape=image_input_shape, name='image_input2')
image_input3 = Input(shape=image_input_shape, name='image_input3')

# Define CNN layers for image inputs
conv1_1 = Conv2D(32, (3, 3), activation='relu', padding='same')(image_input1)
maxpool1_1 = MaxPooling2D(pool_size=(2, 2))(conv1_1)

conv1_2 = Conv2D(32, (3, 3), activation='relu', padding='same')(image_input2)
maxpool1_2 = MaxPooling2D(pool_size=(2, 2))(conv1_2)

conv1_3 = Conv2D(32, (3, 3), activation='relu', padding='same')(image_input3)
maxpool1_3 = MaxPooling2D(pool_size=(2, 2))(conv1_3)

flatten1_1 = Flatten()(maxpool1_1)
flatten1_2 = Flatten()(maxpool1_2)
flatten1_3 = Flatten()(maxpool1_3)

# Concatenate flattened image features
concatenated_images = Concatenate()([flatten1_1, flatten1_2, flatten1_3])

# Define categorical input
occasion_input = Input(shape=(num_occasions,), name='occasion_input')

# Concatenate image features and categorical input
combined_input = Concatenate()([concatenated_images, occasion_input])

# Fully connected layers with BatchNormalization and Dropout
dense1 = Dense(256, activation='relu')(combined_input)
batchnorm1 = BatchNormalization()(dense1)
dropout1 = Dropout(0.5)(batchnorm1)

dense2 = Dense(128, activation='relu')(dropout1)
batchnorm2 = BatchNormalization()(dense2)
dropout2 = Dropout(0.5)(batchnorm2)

output = Dense(1, activation='sigmoid', name='output')(dropout2)  # Linear activation for a regression task

# Create the model
model = Model(inputs=[image_input1, image_input2, image_input3, occasion_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')  # Using mean squared error for regression

# Print model summary
model.summary()
```

```python
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam 
 

from tensorflow.keras.optimizers import SGD
# Compile the model
model.compile(optimizer=SGD(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])




```

```python
# Train the model
model.fit(
     [pixels_x1, pixels_x2,pixels_x3,occasion],
    rating,
    batch_size=10,
    epochs=10
   )
```
